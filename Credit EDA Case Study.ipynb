## 3. Data Quality Check and Missing Values

```python
# Checking percentage of null values in each column

round(100*(appdata.isnull().sum()/appdata.shape[0]), 2).sort_values(ascending=False)
```

```python
# Removing columns with high percentage of missing values (50%)

appdata=appdata.loc[:, round(100*(appdata.isnull().sum()/appdata.shape[0]), 2) <50]
```

```python
# Checking if any columns with high percentage of missing values are still left
round(100*(appdata.isnull().sum()/len(appdata.index)), 2).sort_values(ascending=False)
```

```python
# Removing columns with high percentage of missing values (45%) and checking if any columns with high percentage of missing values still remain

appdata=appdata.loc[:, round(100*(appdata.isnull().sum()/len(appdata.index)), 2) <45]
round(100*(appdata.isnull().sum()/len(appdata.index)), 2).sort_values(ascending=False)
```

```python
appdata.shape
```

### Correlation

#### Matrix plot for Target 0

```python
corr_target0 = target0[columns_for_corr]
corr_mat0 = corr_target0.corr()
plt.figure(figsize=(12,10))
sns.heatmap(corr_mat0, annot=True)
plt.title("Correlation Matrix for Target 0 Columns")
plt.show()
```

```python
# Top 10 correlations for Target 0
corr_mat0 = corr_target0.corr()
corr = corr_mat0.where(np.triu(np.ones(corr_mat0.shape), k=1).astype(np.bool))
corr = corr.abs().unstack().reset_index()
corr.columns = ['VAR1', 'VAR2', 'CORR']
corr.dropna(subset=['CORR'], inplace=True)
corr.sort_values(by='CORR', ascending=False).head(10)
```

```python
#Top 10 correlations for Target 1

corr_mat1 = corr_target1.corr()
corr = corr_mat1.where(np.triu(np.ones(corr_mat1.shape), k=1).astype(np.bool))
corr = corr.abs().unstack().reset_index()
corr.columns = ['VAR1', 'VAR2', 'CORR']
corr.dropna(subset=['CORR'], inplace=True)
corr.sort_values(by='CORR', ascending=False).head(10)
```
